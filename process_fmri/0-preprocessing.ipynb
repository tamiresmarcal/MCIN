### Pre processing options
- motion_correction fsl: mcflirt
- reslice_image fsl: flirt
- interpolate_data fsl: fslmaths
- interpolate_data nilearn: clean_img, smooth_img
    

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.spatial.distance import cdist
from scipy.stats import pearsonr, zscore
import nibabel as nib
from nilearn import datasets, input_data, plotting, connectome, image
import os
import subprocess
```

```python
def run_fsl_command(command):
    """Utility function to run FSL commands."""
    print(f"Running command: {command}")
    process = subprocess.Popen(command, shell=True, stdout=subprocess.PIPE, stderr=subprocess.PIPE)
    stdout, stderr = process.communicate()
    if process.returncode != 0:
       error_message = stderr.decode('utf-8')
       print(f"Error: {error_message}")
       raise RuntimeError(f"Command failed with error: {error_message}")
    return stdout.decode('utf-8')

def motion_correction(input_image, output_image):
    """Correct for head motion during the scan."""
    command = f"mcflirt -in {input_image} -out {output_image} -plots"
    run_fsl_command(command)
    print(f"Function motion_correction done. Output image saved as {output_image}")

def reslice_image(input_image, output_image, voxel_size):
    """Reslice the images to a standard orientation and voxel size."""
    command = f"flirt -in {input_image} -ref {input_image} -out {output_image} -applyisoxfm {voxel_size}"
    run_fsl_command(command)
    print(f"Function reslice_image done. Output image saved as {output_image}")

def interpolate_data(input_image, output_image):
    """Interpolate the data to fill in missing or corrupted values."""
    command = f"fslmaths {input_image} -interp sinc {output_image}"
    run_fsl_command(command)

def smooth_data(input_image, output_image, fwhm):
    """Smooth the data using a Gaussian kernel."""
    command = f"fslmaths {input_image} -s {fwhm} {output_image}"
    run_fsl_command(command)        

def interpolate_data_nilearn(input_image, output_image):
    """Interpolate the data to fill in missing or corrupted values using nilearn's clean_img.
        https://nilearn.github.io/dev/modules/generated/nilearn.image.clean_img.html
    """
    img = nib.load(input_image)
    
    # Apply cleaning to the image
    cleaned_img = image.clean_img(img, detrend=True, standardize=True, low_pass=None, high_pass=None, t_r=None)
    
    # Save the cleaned image
    nib.save(cleaned_img, output_image)
    print(f"Function interpolate_data_nilearn done. Output image saved as {output_image}")

# Path
input_image = '/home/tamires/projects/rpp-aevans-ab/tamires/data/fmri_datasets/ds002837/sub-1/func/sub-1_task-500daysofsummer_run-01_bold.nii.gz'
motion_corrected_image = "/home/tamires/projects/rpp-aevans-ab/tamires/data/fmri_derived/temp/motion_corrected.nii"
resliced_image = "/home/tamires/projects/rpp-aevans-ab/tamires/data/fmri_derived/temp/resliced.nii"
interpolated_image = "/home/tamires/projects/rpp-aevans-ab/tamires/data/fmri_derived/temp/interpolated.nii"

# Run preprocessing steps
motion_correction(input_image, motion_corrected_image)
reslice_image(motion_corrected_image, resliced_image, voxel_size = 2) # voxel size in mm
interpolate_data_nilearn(resliced_image, interpolated_image)
```

```python
# Original image
fmri_img = image.load_img(input_image)
plotting.plot_img(fmri_img.slicer[..., 0], title="original")  # Visualize the first volume
plt.show()

# Motion corrected image
fmri_img = image.load_img(motion_corrected_image)
plotting.plot_img(fmri_img.slicer[..., 0], title="motion corrected")  # Visualize the first volume
plt.show()

#Resliced image
fmri_img = image.load_img(resliced_image)
plotting.plot_img(fmri_img.slicer[..., 0], title="resliced")  # Visualize the first volume
plt.show()

# Clean image
fmri_img = image.clean_img(fmri_img, detrend=True, standardize=False)
plotting.plot_img(fmri_img.slicer[..., 0], title="clean")  # Visualize the first volume
plt.show()

# Smooth image
fmri_img = image.smooth_img(fmri_img, 2.0)
plotting.plot_img(fmri_img.slicer[..., 0], title="smooth")  # Visualize the first volume
plt.show()
```
