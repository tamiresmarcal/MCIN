# Functional Connectivity Basic

- Correlation Matrix:
The correlation matrix is a square matrix where each element (i, j) represents the Pearson correlation coefficient between the time series of region i and region j.
Values range from -1 to 1:
1: Perfect positive correlation (regions are highly synchronized).
0: No correlation (regions are not synchronized).
-1: Perfect negative correlation (regions are inversely synchronized).

- Diagonal Elements: 
The diagonal elements of the matrix are always 1 because they represent the correlation of each region with itself.

- Off-Diagonal Elements:
These elements show the functional connectivity between different regions.
High positive values indicate strong functional connectivity.
High negative values indicate strong inverse connectivity.

- Visualization:
The heatmap visualization helps in identifying clusters of regions that are functionally connected.
Warm colors (e.g., red) indicate positive correlations.
Cool colors (e.g., blue) indicate negative correlations.

- Regions of Interest (ROIs):
If you used an atlas, each row and column corresponds to a specific brain region defined by the atlas.
You can map these indices back to the brain regions to understand which areas are functionally connected.

```python
import nibabel as nib
from nilearn import datasets, input_data, plotting, connectome
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Specify the path to your fMRI data
fmri_path = '/home/tamires/projects/rpp-aevans-ab/tamires/data/fmri_datasets/ds002837/sub-1/func/sub-1_task-500daysofsummer_run-01_bold.nii.gz'
```

```python
# Load fMRI data
fmri_img = nib.load(fmri_path)

# Use a predefined atlas from Nilearn (e.g., Harvard-Oxford atlas)
atlas = datasets.fetch_atlas_harvard_oxford('cort-maxprob-thr25-1mm')
atlas_img = atlas.maps
labels = atlas.labels
print(labels)

# Define the masker to extract time series from ROIs
masker = input_data.NiftiLabelsMasker(labels_img=atlas_img, standardize=True)

# Extract time series data
time_series = masker.fit_transform(fmri_img)
print(time_series.shape)
```

```python
fig = plt.figure()
plt.plot(time_series[:,0])
plt.plot(time_series[:,1])
plt.plot(time_series[:,2])
fig.set_size_inches(15, 5)
legend = [labels[0], labels[1], labels[2]]
plt.legend(legend)
plt.show()
```

```python
fig = plt.figure()
plt.scatter(time_series[:,0], time_series[:,1])
plt.xlabel('Time Series of '+labels[0])
plt.ylabel('Time Series of '+labels[1])
plt.title('Correlation between '+labels[0]+' and '+labels[1])
sns.regplot(x=time_series[:,0], y=time_series[:,1], color='red')
fig.set_size_inches(5, 5)
correlation = np.corrcoef(time_series[:,0], time_series[:,1])[0, 1]
plt.text(0.95, 0.05, 'Pearson Correlation: {:.2f}'.format(correlation), transform=plt.gca().transAxes, ha='right')
plt.show()
```

```python
# Compute the correlation matrix
correlation_measure = connectome.ConnectivityMeasure(kind='correlation')
correlation_matrix = correlation_measure.fit_transform([time_series])[0]

# Ensure the number of labels matches the number of regions
num_regions = correlation_matrix.shape[0]
if len(labels) > num_regions:
    labels = labels[:num_regions]

# Plot the correlation matrix with labels
fig, ax = plt.subplots(figsize=(12, 10))
plotting.plot_matrix(correlation_matrix, 
                     figure=fig, 
                     labels=labels, 
                     vmax=1, 
                     vmin=-1, 
                     title='Functional Connectivity Matrix')
plt.show()

# Visualize the connectome as a graph
fig = plt.figure()
coords = plotting.find_parcellation_cut_coords(labels_img=atlas_img)
plotting.plot_connectome(correlation_matrix, 
                         coords, 
                         edge_threshold="95%", 
                         title='Functional Connectome', 
                         node_color='skyblue', 
                         edge_cmap='coolwarm')
fig.set_size_inches(15, 10)
plt.show()
```
